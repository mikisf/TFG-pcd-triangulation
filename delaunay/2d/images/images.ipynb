```python
import matplotlib.pyplot as plt
import numpy as np

points = [(1, 0), (3, 3), (4, 2), (5, 0), (5, 1), (5, 3)]
edges = [
    (points[2], points[1]),
    (points[2], points[3]),
    (points[3], points[1]),
    (points[4], points[5]),
    (points[5], points[1]),
    (points[2], points[5]),
    (points[4], points[2]),
    (points[3], points[4]),
    (points[0], points[3]),
]

fig, ax = plt.subplots(1, 1, figsize=(6, 4.3))


def draw_arc(ax, i, j, k, arc_radius):
    p0 = np.array(points[i])
    p1 = np.array(points[j])
    p2 = np.array(points[k])

    v1 = p0 - p2
    v2 = p1 - p2
    v1_u = v1 / np.linalg.norm(v1)
    v2_u = v2 / np.linalg.norm(v2)

    angle1 = np.arctan2(v1_u[1], v1_u[0])
    angle2 = np.arctan2(v2_u[1], v2_u[0])

    arc = np.linspace(angle1, angle2, 100)
    arc_x = p2[0] + arc_radius * np.cos(arc)
    arc_y = p2[1] + arc_radius * np.sin(arc)

    ax.plot(arc_x, arc_y, "r-", linewidth=2)


# Plot points
for idx, point in enumerate(points):
    ax.plot(point[0], point[1], "ko", markersize=10)

# Plot edges
for edge in edges:
    p1, p2 = edge
    xs = [p1[0], p2[0]]
    ys = [p1[1], p2[1]]
    ax.plot(xs, ys, "ko-")

draw_arc(ax, 2, 0, 3, 1)
draw_arc(ax, 1, 0, 3, 1.25)
draw_arc(ax, 5, 0, 3, 0.75)

# Finish plot
ax.set_aspect("equal")
ax.axis("off")
plt.xlim(-0.1 + 1, 5.1)
plt.ylim(-0.1, 3.1)
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
fig.patch.set_facecolor("white")
# plt.savefig(f"images/Arcs.png")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

points = [(1, 0), (3, 3), (4, 2), (5, 0), (5, 1), (5, 3)]
edges = [
    (points[2], points[1]),
    (points[2], points[3]),
    (points[3], points[1]),
    (points[4], points[5]),
    (points[5], points[1]),
    (points[2], points[5]),
    (points[4], points[2]),
    (points[3], points[4]),
    (points[0], points[3]),
]

fig, ax = plt.subplots(1, 1, figsize=(6, 4.3))


def draw_circle(ax, i, j, k):
    p0 = np.array(points[i])
    p1 = np.array(points[j])
    p2 = np.array(points[k])

    # Midpoints of two sides
    d = (p0 + p1) / 2
    e = (p1 + p2) / 2

    # Direction vectors
    ab = p1 - p0
    bc = p2 - p1

    # Perpendicular vectors
    ab_perp = np.array([-ab[1], ab[0]])
    bc_perp = np.array([-bc[1], bc[0]])

    # Solve for intersection of perpendicular bisectors
    M = np.vstack([ab_perp, -bc_perp]).T
    b = e - d

    sol = np.linalg.solve(M, b)
    center = d + sol[0] * ab_perp
    radius = np.linalg.norm(center - p0)

    # Draw the circumcircle
    circle = plt.Circle(center, radius, color="red", fill=False, linestyle="--", linewidth=2)
    ax.add_patch(circle)


# Plot points
for idx, point in enumerate(points):
    ax.plot(point[0], point[1], "ko", markersize=10)

# Plot edges
for edge in edges:
    p1, p2 = edge
    xs = [p1[0], p2[0]]
    ys = [p1[1], p2[1]]
    ax.plot(xs, ys, "ko-")

# Draw circle through points 0, 1, and 3
draw_circle(ax, 1, 0, 3)

# Finish plot
ax.set_aspect("equal")
ax.axis("off")
plt.xlim(0.7, 5.3)
plt.ylim(-0.1, 3.1)
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
fig.patch.set_facecolor("white")
# plt.savefig(f"images/Circle1.png")

plt.show()
```
